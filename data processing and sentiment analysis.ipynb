```python
import pandas as pd
df = pd.read_excel('Raw Unstructured data.xlsx')
display(df.head())
```

```python
for col in df:
    print(col)
```

```python
final_df = df[['date','user/id','user/name','profilename','postid','text']].copy()
display(final_df.head())
```

```python
import numpy as np

mask = final_df['postid'].isna()

final_df.loc[mask, 'postid'] = [
    f"PD{str(i+1).zfill(4)}" 
    for i in range(mask.sum())
]
display(final_df.head(2))
```

```python
final_df = final_df.rename(columns={
    'user/id': 'CUSTOMER_ID',
    'profilename': 'CUSTOMER_NAME',
    'postid': 'PRODUCT_ID',
    'text': 'TEXT'
})
display(final_df.head())
```

```python
finall_df=final_df[['date','CUSTOMER_ID','CUSTOMER_NAME','PRODUCT_ID','TEXT']].copy()
display(finall_df.head())
```

```python
pip install textblob
```

```python
from textblob import TextBlob

# Ensure TEXT column is string
finall_df['TEXT'] = finall_df['TEXT'].astype(str)

# Sentiment Score
finall_df['SENTIMENT_SCORE'] = finall_df['TEXT'].apply(
    lambda x: TextBlob(x).sentiment.polarity
)

# Sentiment Label
def get_label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

finall_df['SENTIMENT_LABEL'] = finall_df['SENTIMENT_SCORE'].apply(get_label)

finall_df.head()
```

```python
finall_df.to_excel("sentiment_final_UnstructuredDataset.xlsx", index=False)
```

```python
dff = pd.read_excel('Structured Dataset - Nithya.xlsx')
display(dff.head())
```

```python
dff.shape
```

```python
for col in dff:
    print(col)
```

```python
dff['CustomerID'] = 'CUST' + (dff.index + 1).astype(str).str.zfill(4)
dff['ProductID'] = dff['Food_Items_Consumed'].astype('category').cat.codes
```

```python
def get_sentiment(text):
    analysis = TextBlob(str(text))
    polarity = analysis.sentiment.polarity
    
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

dff['Sentiment_Label'] = dff['Feedback'].apply(get_sentiment)

dff['Sentiment_Score'] = dff['Feedback'].apply(
    lambda x: TextBlob(str(x)).sentiment.polarity
)
```
